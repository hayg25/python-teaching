```python
%matplotlib widget
debut = 180
fin = 250
plt.plot(df1_fr['new_cases_smoothed'][debut:fin])
plt.plot(df1_ch['new_cases_smoothed'][debut:fin])
plt.plot(df1_it['new_cases_smoothed'][debut:fin])
plt.plot(df1_ge['new_cases_smoothed'][debut:fin])
```
