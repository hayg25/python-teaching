### 3-2 Chronogramme de l'évolution de l'épidémie en chine
- Représentation des différents cas (confirmés, décédés, Guerris) 
- La courbe de mortalité semble plate, mais cela est dû à un nombre peu élevé de mortalité par rapport aux autres valeures ()
- Faire un Groupement par date en réalisant la somme de chaque feature
- Faites une représentation sur un même graphe
- Sur un second graphe montrez les evolutions des cas décédés, Guerris, normalisés par les cas confirmés

```python
import matplotlib.pyplot as plt
```

```python
confirmed_by_date_China = []
for d in data_China['Date']:
    confirmed_by_date_China += [data_China["Confirmed"].where(data["Date"] == d).sum() - confirmed_by_date_China[]]
```

```python
deaths_by_date_China = []
for d in data_China['Date']:
    deaths_by_date_China += [data_China["Deaths"].where(data["Date"] == d).sum()]
```

```python
recovered_by_date_China = []
for d in data_China['Date']:
    recovered_by_date_China += [data_China["Recovered"].where(data["Date"] == d).sum()]
```

```python
plt.figure(figsize=(25, 8))


plt.subplot(1,3,1)
plt.plot(data_China['Date'],confirmed_by_date_China)
plt.title('Confirmed')

plt.subplot(1,3,2)
plt.plot(data_China['Date'],deaths_by_date_China, 'r')
plt.title('Deaths')

plt.subplot(1,3,3)
plt.plot(data_China['Date'],recovered_by_date_China, 'g')
plt.title('Recovered')
```
