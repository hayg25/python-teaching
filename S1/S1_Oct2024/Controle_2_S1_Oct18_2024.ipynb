### **Niveau 1 : Variables et fonctions simples**
1. **Exercice 1 : Calculs avec des exceptions**
   - Écrivez une fonction `addition(a, b)` qui prend deux nombres et retourne leur somme. Si l’un des arguments n’est pas un nombre, la fonction doit lever une exception et afficher un message d’erreur approprié.

2. **Exercice 2 : Concaténation améliorée**
   - Écrivez une fonction `concatenation(*args)` qui prend un nombre variable de chaînes de caractères et retourne leur concaténation avec un espace entre chaque chaîne. La fonction doit ignorer les arguments qui ne sont pas des chaînes de caractères.
   Elle doit retourner aussi le nombre de mots ainsi que le nombre total de caractères.

3. **Exercice 3 : Moyenne pondérée**
   - Écrivez une fonction `moyenne_ponderee(notes, coefficients)` qui prend une liste de notes et une liste de coefficients, et retourne la moyenne pondérée. Si les listes ne sont pas de même taille, la fonction doit lever une exception.

### **Niveau 2 : Conditions et boucles complexes**
4. **Exercice 4 : Parité et manipulation des nombres**
   - Écrivez une fonction `analyse_nombre(n)` qui retourne un dictionnaire avec les informations suivantes sur `n` : 
     - Est-il pair ou impair ?
     - Est-il positif, négatif ou nul ?
     - Est-il un carré parfait ?

5. **Exercice 5 : Comparaison de plusieurs nombres**
   - Écrivez une fonction `maximum_liste(liste)` qui prend une liste de nombres et retourne le plus grand nombre, sans utiliser la fonction intégrée `max()`. Gérer les erreurs si la liste contient des éléments non numériques.

6. **Exercice 6 : Somme des carrés**
   - Écrivez une fonction `somme_des_carres(n)` qui retourne la somme des carrés des nombres de 1 à `n`. Vous pouvez avoir l'option, via une variable booléenne, de faire la somme des carrés des nombres pairs ou impairs. 


7. **Exercice 7 : Boucle `while` et gestion des exceptions**
   - Écrivez une fonction `compte_a_rebours(n)` qui affiche un compte à rebours de `n` jusqu’à 0. Gérer les exceptions si l’entrée n’est pas un nombre entier ou si elle est négative.
   Concernant l'affichage, vous pouver utiliser time.sleep(1) pour attendre 1 seconde entre chaque affichage.

### **Niveau 3 : Conditions imbriquées, boucles et complexité**
8. **Exercice 8 : Catégorie d’âge avancée**
   - Écrivez une fonction `categorie_age(age)` qui attribue des catégories d’âge (Enfant, Adolescent, Adulte, Sénior). Incluez une sous-catégorisation (par exemple : "Adolescent jeune" pour 12-15 ans, "Adulte jeune" pour 18-25 ans, etc.).

9. **Exercice 9 : Factorielle récursive**
   - Écrivez une fonction récursive `factorielle(n)` qui calcule la factorielle d’un entier positif. Ajoutez une gestion d’erreurs si l’entrée n’est pas un nombre entier ou si elle est négative.

10. **Exercice 10 : Nombres parfaits**
    - Écrivez une fonction `nombres_parfaits(n)` qui trouve tous les nombres parfaits jusqu’à `n`. Un nombre parfait est un nombre entier positif qui est égal à la somme de ses diviseurs propres, c'est-à-dire ses diviseurs autres que lui-même. 

    Exemple de nombres parfaits : 
    - 6 (diviseurs propres : 1, 2, 3 ; somme : 6)
    - 28 (diviseurs propres : 1, 2, 4, 7, 14 ; somme : 28)

### **Niveau 4 : Manipulation de listes, dictionnaires et fonctions complexes**
11. **Exercice 11 : Liste ordonnée et triée**
    - Écrivez une fonction `verifie_ordre(liste)` qui retourne `True` si la liste est triée, et `False` sinon. Si la liste contient des éléments non comparables (par exemple, des nombres et des chaînes mélangés), gérez cette situation avec des exceptions.

12. **Exercice 12 : Produit et somme des éléments d’une liste**
    - Écrivez une fonction `produit_somme_liste(liste)` qui retourne à la fois la somme et le produit de tous les éléments d’une liste. Assurez-vous que la fonction gère les listes qui contiennent des éléments non numériques.

13. **Exercice 13 : Amélioration de l’algorithme des nombres premiers**
    - Améliorez la fonction `est_premier(n)` pour la rendre plus efficace. Par exemple, utilisez la méthode du "Crible d’Ératosthène" pour vérifier si un nombre est premier.

14. **Exercice 14 : Recherche d’éléments en doublon**
    - Écrivez une fonction `doublons(liste)` qui retourne une liste contenant tous les éléments qui apparaissent plus d'une fois dans la liste d'origine, sans utiliser de collections `set` ou `Counter`.

### **Niveau 5 : Fonctions avancées et manipulation de fichiers**
15. **Exercice 15 : Inverser avec gestion des caractères spéciaux**
    - Écrivez une fonction `inverse_avance(chaine)` qui prend une chaîne de caractères et retourne cette chaîne inversée, en laissant les caractères spéciaux (comme ponctuation, espaces) à leur place originale.

16. **Exercice 16 : Calculatrice avec des expressions complexes**
    - Écrivez une fonction `calculatrice_avancee(expression)` qui prend une chaîne de caractères représentant une expression mathématique (par exemple, `"3 + 5 * (2 - 1)"`) et retourne le résultat en évaluant cette expression. Utilisez la fonction `eval()` avec prudence pour éviter les risques de sécurité. Par exemple vous pouvez boucler dans la chaine et verifier si vous avez un nombre ou un operateur.

17. **Exercice 17 : Analyse de fichier et statistiques**
    - Écrivez une fonction `analyse_fichier(fichier)` qui prend le nom d’un fichier texte, lit son contenu, et retourne un dictionnaire contenant les informations suivantes :
      - Nombre total de mots.
      - Longueur moyenne des mots.
      - Mot le plus fréquent.
      Pour ce dernier vous pouvez utiliser la fonction `Counter` du module `collections`.
      
      Fichier à utiliser dans data/ : 
      - Ballade.XXVIII.dePisan.txt
      - lesdeuxamants.txt


18. **Exercice 18 : Manipulation avancée de fichiers**
    - Écrivez une fonction `recherche_et_remplacement(fichier, ancien_mot, nouveau_mot)` qui prend le nom d’un fichier, remplace toutes les occurrences de `ancien_mot` par `nouveau_mot`, et enregistre les modifications dans un nouveau fichier. Assurez-vous de gérer les fichiers volumineux et les erreurs potentielles liées à l’ouverture et à l’écriture de fichiers.

    Exemple : fichier à utiliser dans data/ : 
      - Ballade.XXVIII.dePisan.txt
      - lesdeuxamants.txt

### **Niveau 6 : Problèmes plus complexes et algorithmes**
19. **Exercice 19 : Vérification avancée de palindrome**
    - Écrivez une fonction `est_palindrome_avance(chaine)` qui vérifie si une chaîne est un palindrome, mais ignore la casse, les espaces et la ponctuation. Par exemple, `"A man, a plan, a canal, Panama"` doit être reconnu comme un palindrome.

20. **Exercice 20 : Bibliotheque**
    Dans cet exercice, tu devras gérer une petite bibliothèque. Chaque livre est représenté par un dictionnaire avec les clés suivantes :

    - titre (string) : le titre du livre
    - auteur (string) : l'auteur du livre
    - année (int) : l'année de publication
    - disponible (bool) : si le livre est disponible ou emprunté

    Ton objectif est de coder des fonctions qui permettent de :

    - Ajouter un livre à la bibliothèque.
    - Supprimer un livre en fonction de son titre.
    - Chercher un livre par son titre et afficher ses informations.
    - Lister tous les livres disponibles.
    - Rendre un livre.
